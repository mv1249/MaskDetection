```python
import numpy as np
import os
import random
import shutil
from glob import glob
from PIL import Image
```

```python
train_path = [x for x in glob('train/*')]
test_path = [x for x in glob('test/*')]
```

```python
train_path
```

```python
os.mkdir('Training')
os.mkdir('Testing')
```

```python
os.mkdir('Training/not_masked')
os.mkdir('Training/masked')
os.mkdir('Testing/not_masked')
os.mkdir('Testing/masked')
```

```python
mask_list = []
normal_list = []
for i in train_path:
    if 'mask' in i:
        mask_list.append(i)
```

```python
mask_list
```

```python
for i in train_path:
    if i not in mask_list:
        normal_list.append(i)
```

```python
normal_list
```

```python
mask_list_test = []
normal_list_test = []
for i in test_path:
    if 'mask' in i:
        mask_list_test.append(i)
```

```python
mask_list_test
```

```python
for i in test_path:
    if i not in mask_list_test:
        normal_list_test.append(i)
```

```python
normal_list_test
```

## Appending Images from train folder to training folder...

#### ---> normal_list is the "not masked"  folder

#### ---> mask_list  is the "masked" folder

```python
original_list = []
train_list = []
for i in normal_list:
    img = Image.open(i)
    original_list.append(img)

for i in original_list:
    image = i.resize((224,224))
    train_list.append(image)
    
for (i,new) in enumerate(train_list):
    new.save('{}{}{}'.format('Training/not_masked/',i+1,'.jpg'))
```

```python
original_list = []
train_list = []
for i in mask_list:
    img = Image.open(i)
    original_list.append(img)

for i in original_list:
    image = i.resize((224,224))
    train_list.append(image)
    
for (i,new) in enumerate(train_list):
    new.save('{}{}{}'.format('Training/masked/',i+1,'.jpg'))
```

## Appending Images from test folder to testing folder...

#### ---> normal_list_test is the "not masked"  folder

#### ---> mask_list_test is the "masked" folder

```python
original_list = []
train_list = []
for i in normal_list_test:
    img = Image.open(i)
    original_list.append(img)

    
for i in original_list:
    image = i.resize((224,224))
    train_list.append(image)
    

for (i,new) in enumerate(train_list):
    new.save('{}{}{}'.format('Testing/masked',i+1,'.jpg'))
```
